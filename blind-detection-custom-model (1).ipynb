# Data Processing

```python
df["id_code"] = df["id_code"].apply(lambda x: x + ".png")
```

```python
train_df = df.iloc[:3000,:]
test_df = df.iloc[3000:,:]
```

```python
train_df['diagnosis'] = train_df['diagnosis'].astype('str')
gen = ImageDataGenerator(
    horizontal_flip = True,
    vertical_flip = True,
    shear_range = 0.2,
    zoom_range = 0.2,
    rescale = 1/255.,
)
train_datagen = gen.flow_from_dataframe(
    train_df,
    directory = "/kaggle/input/aptos2019-blindness-detection/train_images",
    batch_size = 32,
    target_size = (224,224),
    seed = 42,
    x_col = 'id_code',
    y_col = 'diagnosis',
    class_mode = 'categorical'
)
```

```python
test_df['diagnosis'] = test_df['diagnosis'].astype('str')
gen = ImageDataGenerator(
    rescale = 1/255.,
)
test_datagen = gen.flow_from_dataframe(
    test_df,
    directory="/kaggle/input/aptos2019-blindness-detection/train_images",
    batch_size = 32,
    target_size = (224,224),
    seed = 42,
    x_col = 'id_code',
    y_col = 'diagnosis',
    class_mode = 'categorical'
)
```

# Modelling

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, GlobalAveragePooling2D

model = Sequential()

# 1st Conv Layer
model.add(Conv2D(32, (3,3), activation='elu', padding='same', input_shape=(224,224,3)))
model.add(MaxPooling2D(pool_size=(2,2)))

# 2nd Conv Layer
model.add(Conv2D(64, (3,3), activation='elu', padding='same'))
model.add(MaxPooling2D(pool_size=(2,2)))

# 3rd Conv Layer
model.add(Conv2D(128, (3,3), activation='elu', padding='same'))
model.add(MaxPooling2D(pool_size=(2,2)))

# 4th Conv Layer
model.add(Conv2D(256, (3,3), activation='elu', padding='same'))
model.add(MaxPooling2D(pool_size=(2,2)))

# 5th Conv Layer
model.add(Conv2D(512, (3,3), activation='elu', padding='same'))
model.add(MaxPooling2D(pool_size=(2,2)))

# Flatten & Fully Connected Layers
model.add(Flatten())
model.add(Dense(256, activation='elu'))
model.add(Dropout(0.3))

# Output Layer for 5 classes
model.add(Dense(5, activation='softmax'))
```

```python
loss = tf.keras.losses.CategoricalCrossentropy(
    label_smoothing = 0.001,
    name = 'categorical_crossentropy'
)
```

```python
optimizer = Adam(learning_rate = 1e-4)
```

```python
model.compile(optimizer = optimizer,loss=loss,metrics= ['categorical_accuracy'])
```

```python
model.summary()
```

# Model Training 

```python
rlrong = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    mode='min',
    min_lr = 1e-5,
    patience = 2,
    verbose=1
)
estop = EarlyStopping(
    monitor = 'val_loss',
    mode= 'min',
    patience = 3,
    verbose = 1,
    restore_best_weights = True
)
```

```python
history = model.fit(train_datagen,epochs = 20,verbose=1,validation_data = test_datagen,callbacks = [rlrong,estop])
```

```python
plt.plot(history.history['loss'],label='loss',color='red')
plt.plot(history.history['val_loss'],label='val loss',color='blue')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(history.history['categorical_accuracy'],label='categorical accuracy',color='red')
plt.plot(history.history['val_categorical_accuracy'],label='val categorical accuracy',color='blue')
plt.legend()
plt.grid()
plt.show()
```
